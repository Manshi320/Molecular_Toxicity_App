**Read the original OASIS dataset from the Excel file.**

```python
oasis_raw = pd.read_excel("OASIS.xlsx")
```

**Display the column names in the original OASIS dataset.**

```python
print(oasis_raw.columns)
```

**Examine original OASIS labels**

```python
oasis_raw['Value.MeanValue'].value_counts(dropna=False)
```

```python
# Keep only labeled compounds
oasis = oasis_raw[['SMILES', 'Value.MeanValue']].copy()

# Remove rows without an assigned label
oasis = oasis.dropna(subset=['Value.MeanValue'])

# Map labels
oasis['Label'] = oasis['Value.MeanValue'].map({
    'Positive': 1,
    'Negative': 0
})

# Keep final columns
oasis = oasis[['SMILES', 'Label']]
oasis['Source'] = 'OASIS'
```

**Build master dataset**

```python
master = pd.concat([oasis, data2, data3], ignore_index=True)

print(master.shape)
```

**Save the master dataset**

```python
master.to_csv("master_raw.csv", index=False)

print("Master dataset saved successfully!")
```
